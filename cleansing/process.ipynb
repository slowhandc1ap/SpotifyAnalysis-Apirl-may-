```python
artistDF = pd.read_csv('featured_Spotify_artist_info.csv')
```

```python
artistDF.head(15)
```

```python
def clean_genres(genres):
    if pd.isna(genres):
        return genres
    # splitting by comma and 
    genre_list = genres.split(',')
    # keep only first word of each genres 
    # first_word = [genre.split()[0] for genre in genres_list]
    # return only first uniqe
    return genre_list[0]

```

```python
artistDF['cleaningGenres'] = artistDF['genres'].apply(clean_genres) 
```
